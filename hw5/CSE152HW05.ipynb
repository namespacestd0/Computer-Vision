```python
import os, sys
import numpy as np
import numpy.linalg as nl
import scipy.linalg as sl
from imageio import imread
import matplotlib.pyplot as plt
from scipy.ndimage.filters import convolve
from scipy.ndimage.filters import gaussian_filter
from sklearn.decomposition import PCA
import cv2 
```

```python
def load_subset(subsets, base_path='yaleBfaces'):
    imgs = []
    labels = []
    sequence = []
    for subset in subsets:
        directory = os.path.join(base_path, "subset" + str(subset))
        files = os.listdir(directory)
        for img in files:
            face = cv2.imread(os.path.join(directory,img), cv2.IMREAD_GRAYSCALE)
            imgs.append(face)
            labels.append(int(img.split('person')[1].split('_')[0]))
            sequence.append(int(img.split('person')[1].split('_')[1].split('.')[0]))
    imgs = np.array(imgs).astype("float64")
    return imgs, labels, sequence
```

```python
def draw_faces(img_list, col=10):
    if len(img_list) < col:
        col = len(img_list)
        row = 1
    else:
        row = int(len(img_list)/col)
    fig = plt.figure(figsize = (15,2+1.25*row))
    for sub_img in range(1,row*col+1):
        ax = fig.add_subplot(row, col, sub_img)
        ax.imshow(img_list[sub_img-1], cmap='gray')
        ax.axis('off')
    plt.show()
```

## Problem 1 (Programing):  Naive Recognition (15 points)
Perform face recognition by comparing the raw pixel values of the images.

```python
# Let subset 0 be the train set
img, label, seq = load_subset([0])

def nn1_score(im, imgs = img, order = None):
    d = np.empty(len(imgs))
    for k in range(len(imgs)):
        d[k] = np.linalg.norm(imgs[k].ravel() - im.ravel(), order)
    return d

# find nearest neighboor of im in set img
def nn1(im, imgs = img, order=None):
    d = nn1_score(im, imgs, order)
    return np.argmin(d)

def find_training_ind(index):
    return np.where(np.array(label)==index)

# find all photos in set img that belongs to person/class #index
def find_training_set(index, img = img):
    ind = find_training_ind(index)
    return np.array(img)[ind].tolist()

# draw all photos in set img that belongs to person/class #index
def draw_training_set(index, img = None):
    draw_faces(find_training_set(index))
    print("Class",index,"Training Set")
    if img is not None:
        d = nn1_score(img)
        d_of_interest = d[find_training_ind(index)]
        print(d_of_interest.astype(int))
        print(int(d_of_interest.mean()), int(np.median(d_of_interest)))
```

### Use the nearest neighbor classifier (1-NN) using the l2 norm (i.e. Euclidean distance).
#### 1. (5 points) Accuracy

```python
# classified all images of set #i with nn1 algorithm
def predict_naive(test_imgs, train_imgs, train_labels, order = None):
    labels = np.empty(len(test_imgs))
    for j in range(len(test_imgs)):
            labels[j] = train_labels[nn1(test_imgs[j],train_imgs,order)]
    return labels

# compare labels of recognition "labels" with reference in set #i
def evaluate(labels, l, i=""): # l for ref-labels
    print("Subset", i, "accuracy: " 
          + str("{:.6}".format(np.count_nonzero((np.array(labels)-l) == 0) / len(labels) * 100) + "%"))

# report classification accuracy on test sets 1 to 4. 
def report_1nn():
    for i in range(1,5): 
        im, lb, _ = load_subset([i])
        labels = predict_naive(im,img,label)
        evaluate(labels, lb, i)
        
report_1nn()
```

## Problem 2 (Programing): k-Nearest Neighbors Recognition (10 points)

Instead of using a single nearest neighbor, sometimes it is useful to consider the consensus (majority vote) of k-nearest neighbours. 

```python
def nn2(im, k, order = None, imgs=img ):
    d = nn1_score(im, imgs, order)
    return np.argsort(d)[:k]
```

### Repeat Part 1.1 using k-nearest neighbor classifier (k-NN).
#### 1. (4 points) Performance 

```python
def predict_knn(im, k, order):
    # already have the results for k = 1 from Part 1.
    if (k==1):
        return predict_naive(im, img, label)
    lab = np.empty(len(im))
    for j in range(len(im)): # photo in test set
        classes = np.zeros(11)
        d_arg = nn2(im[j],k, order)
        for n in range(k):
            l = label[d_arg[n]]
            classes[l] = classes[l] + 1
        lab[j] = np.argmax(classes)
    return lab
```

```python
#Test the performance for each test dataset using k âˆˆ 1, 3, 5. 
def report_knn(order=None):
    for k in range(3):
        print(k*2+1)
        for i in range(1,5): # test set
            im, lb, _ = load_subset([i])
            labels = predict_knn(im,k*2+1,order)
            evaluate(labels, lb, i)

report_knn()
```

## Problem 3 (Programing): Recognition Using Eigenfaces (25 points)

### (5 points) Find Principle Components
Use Principle Component Analysis (PCA) to get eigenvectors.

```python
# input matrix N Ã— dtrainset of vectorized images from subset 0, 
# where N = 70 is the number of training images 
# and d = 2500 is the number of pixels in each training image.
trainset = np.reshape(img,(70, 2500))

def img_mu(trainset):
    d_2500x70 = np.swapaxes(trainset, 0, 1)
    mu_2500x1 = np.mean(d_2500x70, axis=1)[:,np.newaxis]
    return mu_2500x1, d_2500x70

# Return the d Ã— d matrix of eigenvectors W,
# that could be further sliced to be a k x d matrix,
# and a d dimensional vector mu encoding the mean of the training images.
def PCA_EIG(trainset):
    mu_2500x1, _ = img_mu(trainset)
    s = np.zeros((len(trainset[0]),len(trainset[0])))
    for i in range(len(trainset)):
        a = trainset[i][:,np.newaxis]
        b = (a-mu_2500x1)@(a.T-mu_2500x1.T)
        s = s + b/(len(trainset)-1)
    ss, v = nl.eigh(s)
    return np.fliplr(v), mu_2500x1

# alternative implementation
def PCA_SVD(trainset):
    mu_2500x1, d_2500x70 = img_mu(trainset)
    x_minus_mu_2500x70 = d_2500x70 - mu_2500x1
    u, _, _ = nl.svd(x_minus_mu_2500x70)
    return u, mu_2500x1

basis, mu = PCA_EIG(trainset) 
```

### (2 points) Objective
Performing PCA on the training image observations of $b$ pixels can project the observations in $\Bbb{C}^b$ to a set of values of linearly uncorrelated feature points (principal components) in $\Bbb{C}^k$. It reduces the the amount of data and separtes distinct features of the observations.

### (2 points) Reconstruction
Show what each of these images would look like when using only the top k eigenvectors to reconstruct them, for k = 1, 2, 3, 4, 5, etc. 

```python
p = {}
# Select one image per person from subset 0 
# (e.g., the 5 images person01 01.png, person02 01.png, ... , person10 01.png).
for i in range(len(label)):
    if (seq[i] == 1):
        p[label[i]] = img[i]
draw_faces(list(p.values()))
```

```python
def project(signal, basis, mu, start = 0, k = 1):
    if (len(signal)!=1):
        print("Project Function Error: Signal to decompose needs to be 1 x d dimension.")
        return;
    if (len(basis)!=len(signal[0])):
        print("Project Function Error: Basis should be of d x k dimension")
        return;
    a = signal - mu.reshape((1,-1)) # 1 x d
    # choose first k basis
    v = basis[:, start : start + k] # d x k
    # Project each image into a k dimensional space
    b = a @ v                       # 1 x k 
    # project that k dimensional space back into a 2500 dimensional space
    c = v @ b.T                     # d x 1
    # resize that 2500 vector into a 50 Ã— 50 image.
    r = c.reshape((50,50)) + mu.reshape((50,50)) #normalize
    return b.T, c, r
    # k x 1, d x 1, 50 x 50
    
def demon_reconstruction(start=0, basis = basis, mu = mu): 
    plt.gcf().set_size_inches(20,6)
    toDraw=[]
    for i in range(1,11): # 10 persons
        signal = p[i].ravel()[np.newaxis,:]# image 1 x 2500
        for k in range(1,11): # k=1-10
            _, _, r = project(signal, basis, mu, start = start, k = k)
            toDraw.append(r)
    draw_faces(toDraw)

demon_reconstruction()
```

### (10 points) Recognition

```python
def projects(s,W,mu,start,k):
    result = []
    # Project images from trainset onto the space spanned by the first k eigenvectors 
    for i in range(len(s)):
        kx1, _, _ = project(s[i].reshape((1,-1)),W,mu,start=start,k=k)
        result.append(kx1)
    return np.array(result)
    
# return an M dimensional vector testlabels encoding the predicted class label for each test example. 
def eigenTest(trainset,trainlabels,testset,W,mu,k,start=0):
    # Project images from trainset onto the space spanned by the first k eigenvectors 
    reduced_train = projects(trainset,W,mu,start=start,k=k)[:,:,0] # 70 x 10
    # Project images from testset onto the space spanned by the first k eigenvectors
    reduced_test = projects(testset,W,mu,start=start,k=k)[:,:,0] # 70 x 10
    return np.array(predict_naive(reduced_test, reduced_train, label, order = None))
```

```python
def testMissRate(start = 0):
    miss_rate = np.empty((4,20))
    for i in range(4):
        for k in range(20):
            im, lb, _ = load_subset([i+1])
            test_result = eigenTest(img,label,im,basis,mu,k+1,start)
            # fraction of incorrect predicted class labels
            miss_rate[i,k] = 1 - np.count_nonzero((np.array(test_result)-lb) == 0) / len(lb)
    return miss_rate
```

## Problem 4 (Programing): Recognition Using Fisherfaces (20 points)

```python
def class_mu(trainset_reduced, label):
    mu_class = {}
    for k in range(10):
        for i in range(len(trainset_reduced)):
            if (label[i]==k+1):
                w_img = trainset_reduced[i]/7
                if i in mu_class:
                    mu_class[label[i]] = mu_class[label[i]] + w_img
                else:
                    mu_class[label[i]] = w_img    
    return mu_class

# INPUT
# N Ã— d matrix trainset of vectorized images from subset 0, 
# the corresponding class labels trainlabels, 
# and the number of classes c = 10
def fisherTrain(trainset, trainlabel, c = 10):
    N = len(trainset)
    # Compute the mean mu of the training data
    # and use PCA to compute the first N âˆ’ c principal components. Let this be W_PCA.
    W_PCA_full, mu = PCA_SVD(trainset)
    W_PCA = W_PCA_full[:,:(N-c)] # 2500 x 60
    # Use W_PCA to project the training data into a space of dimension (N âˆ’ c)
    trainset_reduced = (trainset - mu[:,0]) @ W_PCA # 70 x 60
    # Compute the between-class scatter matrix S_B on the (N âˆ’ c) dimensional space from the previous space.
    S_B = np.zeros((N-c,N-c)) # 60 x 60
    mu_class = class_mu(trainset_reduced, label)
    mu_reduced = np.mean(trainset_reduced, axis = 0) # 60,
    for i in range(c):
        b_c = (mu_class[i+1]-mu_reduced).reshape((-1,1))
        S_B = S_B + N*(b_c@b_c.T) # 60 x 60
    # Compute the within class scatter matrix S_W 
    # on the (N âˆ’ c) dimensional space from the previous space.
    S_W = np.zeros((N-c,N-c)) # 60 x 60
    for j in range(70):
        dif = (trainset_reduced[j] - mu_class[label[j]]).reshape((-1,1))
        S_W = S_W + dif @ dif.T
    # Compute W_FLD, by solving the generalized eigenvectors 
    w, vr = sl.eigh(S_B, b = S_W) # 60,
    vr = vr/w.reshape((1,-1))
    args_sorted = np.argsort(np.real(w))
    v = vr[:,args_sorted[::-1]]
    # and use the (câˆ’1) largest generalized eigenvalues 
    W_fld = v[-9:] # 9 x 60
    W_pca = W_PCA.T # 60 x 2500
    # The fisher bases will be a W = W_FLD W_PCA, where W is (c âˆ’ 1) Ã— d dimensional,
    # W_FLD is (c âˆ’ 1) Ã— (N âˆ’ c) dimensional, and W_PCA is (N âˆ’ c) Ã— d dimensional.
    W = W_fld@W_pca # 9 x 2500
    return W
```
