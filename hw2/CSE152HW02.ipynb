## Problem 2: Image Formation and Rigid Body Transformations

In  this  problem  we  will  practice  rigid  body  transformations  and  image  formations  through  the projective camera model. The goal will be to photograph four points in the world coordinate frame. 

Here is the setup code, the four points in world coordinate:

```python
point1 = np.array([[-1,-.5,2]]).T
point2 = np.array([[1,-.5,2]]).T
point3 = np.array([[1,.5,2]]).T
point4 = np.array([[-1,.5,2]]).T
points = np.hstack((point1,point2,point3,point4))
```

The translations happen in homogeneous coordiantes, so we need a functions translating them from and to world coordinate. One is already defined earlier.

```python
# convert points from euclidian to homogeneous
def to_homog(points):
    b = np.array([1, 1, 1, 1]) #how to do it automatically by size
    points_homog = np.vstack((points,b))
    return points_homog
```

The intrinsic transofrmation matrix, also known as camera calibration matrix, is defined by focal length.<br>
The extrinsic transofrmation matrix is defined by rotation and translation on x,y and z.

```python
def Rt(x,y,z,tx,ty,tz):
    Rx = np.array([[1,0,0],
        [0,math.cos(math.radians(x)),-math.sin(math.radians(x))],
        [0,math.sin(math.radians(x)),math.cos(math.radians(x))]])
    Ry = np.array([[math.cos(math.radians(y)),0,math.sin(math.radians(y))],
        [0,1,0],
        [-math.sin(math.radians(y)),0,math.cos(math.radians(y))]])
    Rz = np.array([[math.cos(math.radians(z)),-math.sin(math.radians(z)),0],
        [math.sin(math.radians(z)),math.cos(math.radians(z)),0],
        [0,0,1]])
    R = np.dot(np.dot(Rx,Ry),Rz)
    T = np.array([[tx,ty,tz]]).T
    return np.vstack((np.hstack((R,T)),np.array([0,0,0,1])))

# expanded already for computation
def K(f):
    k = np.array([
        [f,0,0,0],
        [0,f,0,0],
        [0,0,1,0]])
    return k
```

Using the two matrixes and world coordiantes, the matrix multiplication below computes the corresponding 2D points.

```python
# project 3D euclidian points to 2D euclidian
def project_points(P_int, P_ext, pts):
    points_homog = to_homog(pts)
    result_homog = np.array([[0,0,0]]).T
    for i in range(4):
        step1 = np.dot(P_int, P_ext)
        step2 = np.dot(step1, points_homog[:,i:i+1])
        result_homog = np.hstack((result_homog, step2))
    pts_2d=from_homog(result_homog[:,1:])
    return pts_2d

# print parameters
def printM(n,i,e):
    print("Camera ", n)
    print("Intrinsic:\n",i[:,0:3])
    print("Extrinsic:\n",e)
    return

def camera1():
    P_ext = Rt(0,0,0,0,0,0)
    P_int_proj = K(1)
    printM(1,P_int_proj, P_ext)
    return P_int_proj, P_ext

def camera2():
    P_ext = Rt(0,0,0,0,0,1)
    P_int_proj = K(1)
    printM(2,P_int_proj, P_ext)
    return P_int_proj, P_ext

def camera3():
    P_ext = Rt(0,60,30,0,0,1)
    P_int_proj = K(1)
    printM(3,P_int_proj, P_ext)
    return P_int_proj, P_ext

def camera4():
    P_ext = Rt(0,60,30,0,0,13)
    P_int_proj = K(5)
    printM(4,P_int_proj, P_ext)
    return P_int_proj, P_ext
```

In our example, the four different camera settings result in these different images.<br>
1\. [No rigid body transformation]. Focal  length  =  1. The  optical  axis  of  the  camera  is aligned with the z-axis.

```python
i = 0
P_int_proj, P_ext = camera1()
plot_points(project_points(P_int_proj, P_ext, points), title='Camera %d Projective'%(i+1), axis=[-.6,2.5,-.6,1])
plt.show()
```

2\. [Translation]. $\boldsymbol{t} = [\text{0 0 1}]^T$. The optical axis of the camera is aligned with the z-axis. By looking at the scale marking, the rectangle is smaller because it is further away after translation.

```python
i = 1
P_int_proj, P_ext = camera2()
plot_points(project_points(P_int_proj, P_ext, points), title='Camera %d Projective'%(i+1), axis=[-.6,2.5,-.6,1])
plt.show()
```

3\. [Translation and Rotation]. Focal length = 1. $\text{}R$ encodes a 30 degrees around the z-axis and then 60 degrees around the y-axis. $\boldsymbol{t} = [\text{0 0 1}]^T$. Obvious deshape is observed when axial roataions are performed because the distance of points on the shape to the camera is different and perspective distortion is introduced.

```python
i = 2
P_int_proj, P_ext = camera3()
plot_points(project_points(P_int_proj, P_ext, points), title='Camera %d Projective'%(i+1), axis=[-.6,2.5,-.6,1])
plt.show()
```

4\. [Translation and Rotation, long distance]. Focal length = 5. $\text{}R$ encodes a 30 degrees around the z-axis and then 60 degrees around the y-axis. $\boldsymbol{t} = [\text{0 0 13}]^T$. The distortion is less obvious in longer distance.

```python
i = 3
P_int_proj, P_ext = camera4()
plot_points(project_points(P_int_proj, P_ext, points), title='Camera %d Projective'%(i+1), axis=[-.6,2.5,-.6,1])
plt.show()
```

$\textbf{3.3 Surface normals [8 pts]}$

Calculate the surface normals normalized to unit length and display in different resolutions. The surface normals are given by 
\begin{eqnarray}
[-\frac{\delta f}{\delta x}, -\frac{\delta f}{\delta y}, 1].
\end{eqnarray}

```python
def mygradient(test):
    myx = np.empty(test.shape)
    myy = np.empty(test.shape)
    for y in range(test.shape[1]):
        myx[0,y] = test[1,y] - test[0,y]
        for x in range(1,test.shape[0]-1):
            myx[x,y] = (test[x+1,y] - test[x-1,y])/2
        myx[test.shape[0]-1,y] = test[test.shape[0]-1,y] - test[test.shape[0]-2,y]
    for x in range(test.shape[0]):
        myy[x,0] = test[x,1] - test[x,0]
        for y in range(1,test.shape[1]-1):
            myy[x,y] = (test[x,y+1] - test[x,y-1])/2
        myy[x,test.shape[1]-1] = test[x,test.shape[1]-1] - test[x,test.shape[1]-2]
    return (myx,myy)

# (x, y) = np.gradient(heightmap)
(x, y) = mygradient(heightmap)

n_3d = np.dstack((-x, -y, np.ones(x.shape)))
n = n_3d / np.expand_dims(np.linalg.norm(n_3d, axis = 2), axis = 2)

fig = plt.figure()
fig.set_size_inches(7, 4)
ax0 = fig.add_subplot(111, projection='3d')
ax0.set_title("unit length normal vector",y=-0.2)

Xv = np.arange(0, 188, 5)
Yv = np.arange(0, 126, 5)
Yv, Xv, Zv = np.meshgrid(Yv, Xv, 1)
Zv = heightmap[Xv, Yv]
U = n[:,:,0][Xv,Yv]
V = n[:,:,1][Xv,Yv]
W = n[:,:,2][Xv,Yv]

ax0.quiver(Xv,Yv,Zv,U,V,W,length=12)
ax0.set_xlabel('image x axis'); ax0.set_ylabel('image y axis');

plt.show()
```

For each of the two albedos, render three images. One for each of the two light sources, and one for both light-sources combined. Display these in a $2 \times 3$ subplot figure with titles. The general image formation equation is consists of the following 1D items:

* $a(x,y)$ is the albedo for pixel $(x, y)$, <br>

all of above are known as of now. $n$D items include:

$d(x,y)$ the distance to the light,

```python
def D(x, y, ls):
    v = np.array([x, y, heightmap[x, y]]) - light_source[ls]
    return np.linalg.norm(v)

d = np.empty((3,) + albedo.shape)
for x in range(albedo.shape[0]):
    for y in range(albedo.shape[1]):
        for k in range(2):
            d[k, x, y] = D(x, y, k)
```

$\hat{s}(x,y)$ the light source direction vector,  

```python
def S(x, y, ls):
    v = light_source[ls] - np.array([x, y, heightmap[x, y]])
    return v/np.linalg.norm(v)

s = np.empty((3,) + n.shape)
for x in range(albedo.shape[0]):
    for y in range(albedo.shape[1]):
        for k in range(2):
            s[k, x, y] = S(x, y, k)
```

And eventually $I = a(x,y) \hat{\boldsymbol{n}}(x, y)^\top \hat{\boldsymbol{s}}(x, y)\frac{s_0}{( d(x,y) )^{2}}$

```python
# ls is the light source index 0 or 1
# a is the albedo source index 0 or 1, 1 for uniform
def I(x, y, ls, a):
    if (not a):
        return albedo[x,y] * np.dot(n[x,y],S(x,y,ls)) / (D(x,y,ls)**2)
    else:
        return uniform_albedo[x,y] * np.dot(n[x,y],S(x,y,ls)) / (D(x,y,ls)**2)
    
Images = np.empty((6,) + albedo.shape)

for i in range(6):
    if (i%3 == 2):
        Images[i] = Images[i-1] + Images[i-2]
    else:
        for x in range(albedo.shape[0]):
            for y in range(albedo.shape[1]):
                Images[i, x, y] = I(x, y, i%3, math.floor(i/3))
```
